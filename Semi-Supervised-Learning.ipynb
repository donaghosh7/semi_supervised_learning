```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.linear_model import LogisticRegression
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
import time
```

### <span style=font-family:Times New Roman> Loading, Pre-processing and Visualizing the Dataset</span>

```python
fashion_mnist=keras.datasets.fashion_mnist
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
print("X_train shape:" + str(X_train.shape))
print("y_train shape:" + str(y_train.shape))
print("X_test shape:" + str(X_test.shape))
print("y_test shape:" + str(y_test.shape))
```

```python
X_train = X_train/255.0
X_test = X_test/255.0
```

```python
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
"Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

```python
plt.figure(figsize=(10,10))
for i in range(16):
    plt.subplot(4,4,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_train[i], cmap=plt.cm.binary)
    plt.title("Image label is: {}".format(y_train[i]))
plt.show()
```

```python
X_train = X_train.reshape(60000,-1)
X_test = X_test.reshape(10000,-1)
```

### <span style=font-family:Times New Roman>Applied Logistic Regression on the Original Fully Labelled Dataset</span>

```python
start_time = time.time()
log_reg = LogisticRegression(multi_class="ovr", solver="saga", max_iter = 1000, random_state=42)
log_reg.fit(X_train, y_train)
end_time = time.time()
print(log_reg.score(X_test, y_test))
print(f"Total time taken = {end_time-start_time}")
```

### <span style=font-family:Times New Roman>Applied Logistic Regression After Clustering the Training Dataset</span>

```python
k=[100,200,300]
start_time = time.time()
for i in k:
    pipeline = Pipeline([("kmeans", KMeans(n_clusters=i, random_state=42)),
    ("log_reg", LogisticRegression(multi_class="ovr", solver="saga", max_iter=5000, random_state=42))])
    pipeline.fit(X_train, y_train)
    print(f"No. of cluster = {i}, Accuracy = {pipeline.score(X_test, y_test)}")
end_time = time.time()
print(f"Total time taken = {end_time-start_time}")    
```

### <span style=font-family:Times New Roman>Taking Random No. of Labelled Instances</span>

```python
k=[1000,2000,3000]
start_time=time.time()
for i in k:
    log_reg = LogisticRegression(multi_class="ovr", solver="saga", max_iter=5000, random_state=42)
    log_reg.fit(X_train[:i], y_train[:i])
    print(f"Labelled Instance = {i}, Accuracy = {log_reg.score(X_test, y_test)}")
end_time = time.time()
print(f"Total time taken = {end_time-start_time}")
```

### <span style=font-family:Times New Roman>Using Clustering for Semi-Supervised Learning</span>

```python
start_time=time.time()
kmeans = KMeans(n_clusters=3000, random_state=42)
X_items_dist = kmeans.fit_transform(X_train)
representative_item_idx = np.argmin(X_items_dist, axis=0)
X_representative_items = X_train[representative_item_idx]
end_time = time.time()
print(f"Total time taken = {end_time-start_time}")
```

```python
y_train[representative_item_idx]
```

```python
y_representative_items = np.array(y_train[representative_item_idx])
```

```python
start_time=time.time()
log_reg = LogisticRegression(multi_class="ovr", solver="saga", max_iter=1000, random_state=42)
log_reg.fit(X_representative_items, y_representative_items)
end_time = time.time()
print(log_reg.score(X_test, y_test))
print(f"Total time taken = {end_time-start_time}")
```

### <span style=font-family:Times New Roman>Clustering and Propagating the Labels to other Instances in the Same Cluster</span>

```python
y_train_propagated = np.empty(len(X_train), dtype=np.int32)
for i in range(3000):
    y_train_propagated[kmeans.labels_==i] = y_representative_items[i]
```

```python
start_time=time.time()
log_reg = LogisticRegression(multi_class="ovr", solver="saga", max_iter=1000, random_state=42)
log_reg.fit(X_train, y_train_propagated)
end_time = time.time()
print(log_reg.score(X_test,y_test))
print(f"Total time taken = {end_time-start_time}")
```

### <span style=font-family:Times New Roman>Propagating the Labels to Some Certain Percentile Closest to the Centroid</span>

```python
percentile_closest = 25

X_cluster_dist = X_items_dist[np.arange(len(X_train)), kmeans.labels_]
for i in range(3000):
    in_cluster = (kmeans.labels_ == i)
    cluster_dist = X_cluster_dist[in_cluster]
    cutoff_distance = np.percentile(cluster_dist, percentile_closest)
    above_cutoff = (X_cluster_dist > cutoff_distance)
    X_cluster_dist[in_cluster & above_cutoff] = -1

partially_propagated = (X_cluster_dist != -1)
X_train_partially_propagated = X_train[partially_propagated]
y_train_partially_propagated = y_train_propagated[partially_propagated]
```

```python
start_time = time.time()
log_reg = LogisticRegression(multi_class="ovr", solver="saga", max_iter=1000, random_state=42)
log_reg.fit(X_train_partially_propagated, y_train_partially_propagated)
end_time = time.time()
print(f"Percentile Closest = {percentile_closest}, Accuracy = {log_reg.score(X_test,y_test)}")
print(f"Total time taken = {end_time-start_time}")
```
